### 1) ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creamos un array de 1000 elementos booleanos
arr_booleanos = np.array([True] * 1000, dtype=bool)

# Calculamos el tamaño en bytes
tamaño_bytes = arr_booleanos.nbytes

print(f"Tamaño del array de 1000 booleanos: {tamaño_bytes} bytes")
print("En NumPy, un valor booleano ocupa 1 byte (8 bits).")
```

### 3) Dado el array: [2, 4, 5, 6, 0, 3, 7, 4, 8, 8, 5, 2, 1, 5, 6, 1]

Seleccione con una instrucción el subarray de elementos: [0, 3, 7, 4]
Después, seleccione el subarray de elementos: [2, 5, 8, 5]

```python
# Array original
arr = np.array([2, 4, 5, 6, 0, 3, 7, 4, 8, 8, 5, 2, 1, 5, 6, 1])
print("Array original:")
print(arr)

# Primer subarray: [0, 3, 7, 4] - están en índices 4,5,6,7
subarray1 = arr[4:8]
print(f"\nPrimer subarray (índices 4 a 7): {subarray1}")

# Segundo subarray: [2, 5, 8, 5] - están en índices 0, 2, 8, 10
subarray2 = arr[[0, 2, 8, 10]]
print(f"Segundo subarray (índices 0, 2, 8, 10): {subarray2}")
```

### 4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

diferencia = a - b
media = np.mean(diferencia)

print(f"Array a: {a}")
print(f"Array b: {b}")
print(f"Diferencia (a-b): {diferencia}")
print(f"Media aritmética de la diferencia: {media}")
```

### 5) Tengo valores de coordenadas (x, y) en las columnas del array: [[1.33, 4.5], [30.0, 10.7], [70.2, 0.5]]

Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.

```python
# Array original de coordenadas
coordenadas = np.array([[1.33, 4.5], [30.0, 10.7], [70.2, 0.5]])
print("Array original:")
print(coordenadas)
print(f"Dimensiones originales: {coordenadas.shape}")

# Nueva coordenada a agregar
nueva_coord = np.array([[37.1, -3.6]])

# Agregamos usando vstack (apilamiento vertical)
coordenadas_actualizado = np.vstack([coordenadas, nueva_coord])

print("\nArray actualizado:")
print(coordenadas_actualizado)
print(f"Dimensiones del nuevo array: {coordenadas_actualizado.shape}")
```

### 6) Copie el array del ejercicio anterior. Traspóngalo. Agregue ahora dos nuevos pares de coordenadas: (10.8, 3.0) y (35.8, 12.0).

```python
# Copiamos el array anterior
copia_coordenadas = coordenadas_actualizado.copy()
print("Copia del array:")
print(copia_coordenadas)

# Trasponemos (intercambiamos filas por columnas)
traspuesta = copia_coordenadas.T
print("\nArray traspuesto:")
print(traspuesta)
print(f"Dimensiones después de trasponer: {traspuesta.shape}")

# Agregamos dos nuevos pares
nuevos_pares = np.array([[10.8, 3.0], [35.8, 12.0]])
final = np.vstack([copia_coordenadas, nuevos_pares])

print("\nArray final con nuevos pares agregados:")
print(final)
print(f"Dimensiones finales: {final.shape}")
```

### Cargamos el dataset NucleosPoblacion.csv

```python
# Cargamos el archivo CSV
df = pd.read_csv('NucleosPoblacion.csv')

# Mostramos las primeras filas para ver la estructura
print("Primeras 5 filas del dataset:")
print(df.head())

print("\nInformación general del dataset:")
print(df.info())

print("\nEstadísticas descriptivas básicas:")
print(df.describe())
```

### 1) ¿Cuántos Municipios tienen más de 100000 habitantes?

### 2) Realice una gráfica de barras sobre la población de cada ciudad, ordénela de menor a mayor y responda:

a. ¿Cuál es la segunda ciudad más poblada?
b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
# Ordenamos de menor a mayor población
df_ordenado = df.sort_values('Poblacion', ascending=True)

# Creamos la gráfica de barras
plt.figure(figsize=(20, 8))
plt.bar(range(len(df_ordenado)), df_ordenado['Poblacion'])
plt.xlabel('Municipios')
plt.ylabel('Población')
plt.title('Población de cada municipio (ordenada de menor a mayor)')
plt.xticks(range(0, len(df_ordenado), 50), rotation=90, fontsize=8)
plt.tight_layout()
plt.show()

# a) Segunda ciudad más poblada
top_ciudades = df.nlargest(3, 'Poblacion')
segunda_mas_poblada = top_ciudades.iloc[1]
print(f"a) La segunda ciudad más poblada es: {segunda_mas_poblada['Municipio']} con {segunda_mas_poblada['Poblacion']:,.0f} habitantes")

# b) Posición de Granada en el ranking (orden descendente)
ranking_desc = df.sort_values('Poblacion', ascending=False).reset_index(drop=True)
posicion_granada = ranking_desc[ranking_desc['Municipio'] == 'Granada'].index[0] + 1
print(f"b) Granada ocupa la posición #{posicion_granada} en el ranking de ciudades más pobladas")
```

### 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
# Obtenemos las provincias únicas
provincias = df['CodProvin'].unique()
n_provincias = len(provincias)

# Calculamos cuántas filas y columnas necesitamos para los subplots
n_cols = 4
n_rows = (n_provincias + n_cols - 1) // n_cols

# Creamos los subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
axes = axes.flatten()

# Para cada provincia, creamos un histograma
for i, prov in enumerate(provincias):
    datos_prov = df[df['CodProvin'] == prov]['Poblacion']
    axes[i].hist(datos_prov, bins=20, edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Provincia {prov}')
    axes[i].set_xlabel('Población')
    axes[i].set_ylabel('Frecuencia')

# Ocultamos los subplots vacíos
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.suptitle('Histogramas de población por provincia', y=1.02, fontsize=16)
plt.show()
```

### 7) Seleccione al azar cincuenta municipios diferentes de entre los disponibles en el archivo. Asegúrese de que no se repitan. Luego calcule el promedio de la población y la desviación estándar de estos 50 municipios.

```python
# Fijamos una semilla para reproducibilidad
np.random.seed(42)

# Seleccionamos 50 municipios aleatorios sin repetición
muestra_50 = df.sample(n=50, replace=False, random_state=42)

print("Municipios seleccionados aleatoriamente:")
print(muestra_50[['Municipio', 'Provincia', 'Poblacion']].to_string(index=False))

print("\nEstadísticas de la muestra:")
print(f"Promedio de población: {muestra_50['Poblacion'].mean():,.2f}")
print(f"Desviación estándar: {muestra_50['Poblacion'].std():,.2f}")
```

### 8) Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
# Obtenemos las coordenadas de Madrid
madrid = df[df['Municipio'] == 'Madrid'].iloc[0]
print(f"Coordenadas de Madrid: ({madrid['X']}, {madrid['Y']})")

# Excluimos Madrid del análisis
df_sin_madrid = df[df['Municipio'] != 'Madrid'].copy()

# Calculamos la distancia euclidiana a Madrid
# Distancia = sqrt((x2 - x1)^2 + (y2 - y1)^2)
df_sin_madrid['distancia'] = np.sqrt(
    (df_sin_madrid['X'] - madrid['X'])**2 + 
    (df_sin_madrid['Y'] - madrid['Y'])**2
)

# Encontramos el más cercano y el más lejano
idx_cercano = df_sin_madrid['distancia'].idxmin()
idx_lejano = df_sin_madrid['distancia'].idxmax()

municipio_cercano = df_sin_madrid.loc[idx_cercano]
municipio_lejano = df_sin_madrid.loc[idx_lejano]

print("\n" + "="*60)
print("RESULTADOS:")
print("="*60)
print(f"Municipio MÁS CERCANO a Madrid:")
print(f"  Nombre: {municipio_cercano['Municipio']}")
print(f"  Provincia: {municipio_cercano['Provincia']}")
print(f"  Distancia: {municipio_cercano['distancia']:.4f} grados")
print(f"  Población: {municipio_cercano['Poblacion']:,.0f} habitantes")

print(f"\nMunicipio MÁS LEJANO a Madrid:")
print(f"  Nombre: {municipio_lejano['Municipio']}")
print(f"  Provincia: {municipio_lejano['Provincia']}")
print(f"  Distancia: {municipio_lejano['distancia']:.4f} grados")
print(f"  Población: {municipio_lejano['Poblacion']:,.0f} habitantes")
```
