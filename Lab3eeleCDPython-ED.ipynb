##Ejercicio 1: Convertir número decimal a binario##

```python
def decimal_a_binario(n):
    if n == 0:
        return "0"
    binario = ""
    while n > 0:
        residuo = n % 2
        binario = str(residuo) + binario
        n = n // 2
    return binario

# Ejemplo de uso
numero = 250
print(f"El número {numero} en binario es: {decimal_a_binario(numero)}")
```

##Ejercicio 2: Verificar si un número es primo##

```python
def es_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Ejemplo de uso
print(es_primo(75))  # True
print(es_primo(1033))  # False
```

##Ejercicio 3: Calcular desviación estándar

```python
def desviacion_estandar(lista):
    n = len(lista)
    if n <= 1:
        return 0
    promedio = sum(lista) / n
    suma_cuadrados = 0
    for valor in lista:
        suma_cuadrados += (valor - promedio) ** 2
    varianza = suma_cuadrados / (n - 1)
    return varianza ** 0.5

# Ejemplo de uso
lista = [10, 12, 23, 23, 16, 23, 21, 16]
print(f"Desviación estándar: {desviacion_estandar(lista)}")
```

##Ejercicio 4: Ordenamiento burbuja##

```python
def ordenamiento_burbuja(arr):
    nueva_lista = arr.copy()
    n = len(nueva_lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if nueva_lista[j] > nueva_lista[j + 1]:
                nueva_lista[j], nueva_lista[j + 1] = nueva_lista[j + 1], nueva_lista[j]
    return nueva_lista

def main():
    original = [64, 34, 25, 12, 22, 11, 90]
    ordenada = ordenamiento_burbuja(original)
    print("Original:", original)
    print("Ordenada:", ordenada)

if __name__ == "__main__":
    main()
```

##Ejercicio 5: Convertir tupla de elementos a string##

```python
def tupla_a_string(tupla):
    resultado = ""
    for elemento in tupla:
        resultado += str(elemento)
    return resultado

# Ejemplo
entrada = (10, 20, 40, 5, 70)
print(tupla_a_string(entrada))  # "102040570"
```

##Ejercicio 6: Eliminar tuplas vacías de una lista##

```python
def eliminar_tuplas_vacias(lista):
    return [item for item in lista if item != ()]

# Ejemplo
lista = [(), (), ('X',), ('a', 'b'), ('a', 'b', 'c'), ('d',)]
resultado = eliminar_tuplas_vacias(lista)
print(resultado)  # [('X',), ('a', 'b'), ('a', 'b', 'c'), ('d',)]
```

##Ejercicio 7: Promedio de cada tupla en una tupla de tuplas##

```python
def promedios_tuplas(tuplas):
    promedios = []
    for t in tuplas:
        promedio = sum(t) / len(t)
        promedios.append(promedio)
    return tuple(promedios)

# Ejemplo
entrada = ((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32))
print(promedios_tuplas(entrada))  # (10.5, 44.0, 58.0)
```

##Ejercicio 8: Clase Libro (POO)

```python
class Libro:
    def __init__(self, titulo, autor, paginas):
        self.titulo = titulo
        self.autor = autor
        self.paginas = paginas
        self.leido = False

    def marcar_como_leido(self):
        self.leido = True
        print(f"'{self.titulo}' ha sido marcado como leído.")

    def descripcion(self):
        estado = "Leído" if self.leido else "No leído"
        return f"'{self.titulo}' por {self.autor}, {self.paginas} páginas. Estado: {estado}."

# Ejemplo de uso
def menu_libros():
    print("=== SISTEMA DE GESTIÓN DE LIBROS ===")
    biblioteca = []
    
    while True:
        print("\n--- MENÚ ---")
        print("1. Agregar nuevo libro")
        print("2. Ver todos los libros")
        print("3. Marcar libro como leído")
        print("4. Salir")
        
        opcion = input("Seleccione una opción (1-4): ")
        
        if opcion == "1":
            print("\n--- NUEVO LIBRO ---")
            titulo = input("Título: ")
            autor = input("Autor: ")
            try:
                paginas = int(input("Número de páginas: "))
                libro = Libro(titulo, autor, paginas)
                biblioteca.append(libro)
                print("✅ Libro agregado exitosamente.")
            except ValueError:
                print("❌ Error: Las páginas deben ser un número entero.")
        
        elif opcion == "2":
            print("\n--- MIS LIBROS ---")
            if not biblioteca:
                print("No hay libros en la biblioteca.")
            else:
                for i, libro in enumerate(biblioteca, 1):
                    print(f"{i}. {libro.descripcion()}")
        
        elif opcion == "3":
            print("\n--- MARCAR COMO LEÍDO ---")
            if not biblioteca:
                print("No hay libros en la biblioteca.")
            else:
                for i, libro in enumerate(biblioteca, 1):
                    print(f"{i}. '{libro.titulo}' - {libro.autor}")
                try:
                    indice = int(input("Seleccione el número del libro: ")) - 1
                    if 0 <= indice < len(biblioteca):
                        biblioteca[indice].marcar_como_leido()
                    else:
                        print("❌ Número de libro inválido.")
                except ValueError:
                    print("❌ Error: Ingrese un número válido.")
        
        elif opcion == "4":
            print("¡Hasta pronto!")
            break
        else:
            print("❌ Opción no válida. Intente de nuevo.")

# Ejecutar el menú
menu_libros()
```

##Ejercicio 9: Clase Circulo con menú interactivo##
para crear una clase Circulo con métodos para área y circunferencia usamos la constante math.pi para los cálculos.

```python
import math

class Circulo:
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * (self.radio ** 2)

    def circunferencia(self):
        return 2 * math.pi * self.radio

# Ejemplo de uso
circulo = Circulo(10)
print(f"Área: {circulo.area():.2f}")
print(f"Circunferencia: {circulo.circunferencia():.2f}")
```
